```python
with open('origintext.html', 'r', encoding='utf8') as file:
    html_content = file.read()

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.name)
print(soup.p)
```

```python
unique_classes = set()

for tag in soup.find_all(class_=True):
    class_ = tag['class']
    unique_classes.update(class_)

print(unique_classes)
print(len(unique_classes))
```

```python
all_links = set()
# all_links1 = set()
for anchor in soup.find_all('a'):
    href = anchor.attrs.get("href")
    # href1 = anchor['href']
    if (href != "" and href != None):
        all_links.add(href)
    # if (href1 != "" and href1 != None):
    #     all_links1.add(href1)
print(all_links)
# print(all_links1)

# do not use anchor['href'] to get the attribute directly
# if anchor do not have a href, an error will be returned
```

```python
from url_normalize import url_normalize
base_url = 'http://ai.ruc.edu.cn/'
links = set()
for href in all_links:
    if not href.startswith('http'):
        href = url_normalize(base_url + href)
    links.add(href)
print(links)
```

```python
import jieba
title_words = jieba.lcut(soup.title.string)
print(title_words)

doctitle_words = []
for title in soup.find_all({'h1', 'h2', 'h3'}):
    if title.string:
        temp_words = jieba.lcut(title.string)
        doctitle_words += temp_words
print(doctitle_words)
```

```python
import jieba
parsed_text = "".join(
    [p.string for p in soup.find_all('p') if p.string]
)
if parsed_text:
    body_words = jieba.lcut(parsed_text)
else:
    body_words = []
print(parsed_text)
print(body_words)
```
